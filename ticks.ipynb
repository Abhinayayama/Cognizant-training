```python
import matplotlib.pyplot as plt
products=["pens","books","bags"]
sales=[600,2000,1500]
plt.bar(products,sales)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
products=["pens","books","bags"]
sales_2025=[600,2000,1500]
sales_2022=[500,1500,800]
sales_2024=[300,1000,700]
x=np.arange(len(products))
width=0.27
plt.bar(x-width/1,sales_2025,width,label='2025')
plt.bar(x-width/4,sales_2022,width,label='2022')
plt.bar(x-width/10,sales_2024,width,label='2024')
plt.xlabel("products")
plt.title("sales comparison: 2025 vs 2026")
plt.xticks(x,products)
plt.legend()
plt.show()
          
```

```python
import matplotlib.pyplot as plt
import numpy as np
products=["pens","books","bags"]
online=[600,2000,1500]
offline=[500,1500,800]
wholesale=[200,900,700]
plt.bar(products,online,label='online')
plt.bar(products,offline,bottom=online,label='offline')
plt.bar(products,wholesale,label='wholesale')
bottom_values=[online[i]+offline[i] for i in range(len(products))]
plt.bar(products,wholesale,bottom=bottom_values,label='wholesale')
plt.xlabel("products")
plt.ylabel("Revenue")
plt.title("Revenue of the distrubuted sales")
plt.legend()
plt.xticks(x,products)
plt.show()
```
