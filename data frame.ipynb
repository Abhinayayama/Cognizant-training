```python
import pandas as pd
data={
     "Name":["abhinaya","nithishma","vaishnavi","ananya","pragna","deekshitha"],
     "age":[89,99,92,45,39,79],
     "city":["karimnagar","andaman","africa","pune","choppadhandi","godavarikhani"]
}
details=pd.DataFrame(data)
print(details[details.age>50])
print(details[details.age>50]),(details[details.age<20])
details["Grade"]=["A","B","C","D","E","F"]
print(details)
details["age"]=details["age"] +2
print(details)
```

```python
import pandas as pd
datajoin=pd.DataFrame({
     "Name":["deekshitha","nithishma","anana","vaishnavi","abhinaya","pragna"],
     "ID":[89,99,92,45,39,79]})
print(datajoin)

```

```python
datajoin2=pd.DataFrame({
     "marks":[24,56,90,56,39,19],
     "ID":[89,99,92,45,39,79]})
print(datajoin2)
```

```python
result=pd.merge(datajoin,datajoin2, on="ID")
print(result)
```

```python
abc=pd.merge(datajoin,datajoin, on="ID" , how="inner")
print(abc)
```

```python
leftJoin=pd.merge(datajoin,datajoin2, on="ID" , how="left")
print(leftJoin)
```
