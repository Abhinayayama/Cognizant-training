```python
Customers=list(range(1,200))
sample=Customers[::20]
print("Convenience sample",sample)
```

```python
employee={
          "A":10,
          "B":40,
          "C":50,
          "D":60
}
sample=[name for name,exp in employee.items() if exp>10]
print("Judgment sample: ",sample)
```

```python
networks={
          "A":["B","C"],
          "B":["D"],
          "C":["E"]
}
sample.extend(networks["A"])
sample.extend(networks["B"])
print("snowball sapling:",sample)
```
