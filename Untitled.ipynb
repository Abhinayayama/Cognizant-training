```python
import matplotlib.pyplot as plt
import numpy as np
electronics=["bulbs","sandils","ac's","coolers","fridges"]
profits_2020=[500,1200,3000,1800,3500]
profits_2023=[600,1500,2000,1300,6000]
profits_2026=[1000,3000,3000,2400,7000]
x=np.arange(len(electronics))
width=0.22
plt.bar(x-width/1,profits_2020,width,label='2020')
plt.bar(x-width/1.5,profits_2023,width,label='2023')
plt.bar(x-width/6,profits_2026,width,label='2026')
plt.xlabel("electronics")
plt.title("profits comparions:2020 vs 2023 vs 2026")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

months=["jan","feb","mar","apr","may"]
sales=[390,200,499,700,560]
sns.lineplot(x=months,y=sales)
plt.xlabel=("months")
plt.ylabel=("sales")
plt.title("monthly sales trends")
plt.show()
```

```python
import plotly.express as px
import pandas as pd
df=pd.DataFrame({
    'months':['jan','feb','mar','apr','may'],
    'sales':[500,600,490,307,529]})
fig=px.line(df,x='months',y='sales',markers=True)
fig.show()
```

```python
import plotly.express as px
import pandas as pd
df=pd.DataFrame({
    'students':['abhi','hamsi','lucky','manu'],
    'first year':['absent','present','present','absent'],
    'second year':['present','present','absent','present']})
fig=px.line(df,x='students',y='first year',markers=True)
fig=px.line(df,x='students',y='second year',markers=True)
fig.show()
```
