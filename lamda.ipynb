```python
def square(x):
    return x*x

print(square(7))
```

```python
square=lambda x:x*x
print(square(6))
```

```python
mul=lambda a,b:a*b
print(mul(6,3))
```

```python
check=lambda num:"Even" if num%5==0 else "odd"
print(check(7))
```

```python
students=[("vyshu",50),("sweety",90),("abhi",60)]
result=sorted(students, key=lambda x:x[1])
print(result)
```

```python
numbers = [2,5,8,6,7]
result = list(map(lambda x: x*x,numbers))
print(result)
```

```python
numbers = [1,2,3,4,5]
result=list(filter(lambda x:x%2==0,numbers))
print(result)
```

```python
result=list(filter(lambda x:x%5==3,numbers))
print(result)
```

```python
from functools import reduce
numbers=[2,5,6,4,8]
result=reduce(lambda x,y: x+y,numbers)
print(result)
```
